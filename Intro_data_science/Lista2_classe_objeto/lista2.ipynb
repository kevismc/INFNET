3) Usando o e-book 'Alice’s Adventures in Wonderland' contido em Alice.txt:

(a) Faça um programa que retorne o número total de palavras contidas no e-book;

(b) Um programa que retorne um dicionário contendo a frequência de cada palavra. Por exemplo: se a palavra 'guerra' aparece 10 vezes em um texto que contém 50 palavras, então a frequência da palavra 'guerra' é de 20%.

(c) Retorne as 20 palavras mais frequentes do e-book;

```python
#(a) nao deu certo abrir com while
caminho_arquivo = '/content/drive/MyDrive/Alice.txt'

with open(caminho_arquivo, 'r') as arquivo:
    conteudo = arquivo.read()
    palavras = conteudo.split()
    numero_palavras = len(palavras)

print(f"O número total de palavras no e-book 'Alice's Adventures in Wonderland' é: {numero_palavras}")
#(b: tem uma explicação da monitoria sobre dicionário e função lambda na monitoria dia 11/05 min 18)
#(c)
texto = open('/content/drive/MyDrive/Alice.txt')
d = dict()
for linha in texto:
  linha = linha.strip()
  linha = linha.lower()
  palavras =linha.split(" ")

  for palavra in palavras:
    if palavra in d:
      d[palavra] = d[palavra] + 1
    else:
      d[palavra] = 1

dict_ordenado = dict(sorted(d.items(),key = lambda x: x[1], reverse = True))

contador = 0
for chave in list(dict_ordenado.keys()):
  if contador < 20:
    print(chave,":",(dict_ordenado[chave]/numero_palavras)*100)
    contador +=1
  else:
    break
```

4) Refaça o exercício usando o conceito de Classe e Objeto. A sua Classe deve ser aplicada a qualquer contexto, não somente ao problema acima. De forma que ela seja genérica. Cada método deve representar uma única ação do Objeto e não, especificamente, uma função que executa diversas ações. Pensem em ações atômicas. Boa sorte !

```python
class ProcessadorTexto:
    def __init__(self, arquivo):
        self.arquivo = arquivo
        self.texto = None
        self.palavras = None

    def abrir_arquivo(self):
        with open(self.arquivo, 'r') as file:
            self.texto = file.read()

    def contar_palavras(self):
        self.palavras = self.texto.split()
        return len(self.palavras)

    def calcular_frequencia_palavras(self):
        frequencia = {}
        for palavra in self.palavras:
            if palavra in frequencia:
                frequencia[palavra] += 1
            else:
                frequencia[palavra] = 1
        return frequencia

    def obter_palavras_mais_frequentes(self, quantidade):
        frequencia_palavras = self.calcular_frequencia_palavras()
        palavras_mais_frequentes = sorted(frequencia_palavras, key=frequencia_palavras.get, reverse=True)[:quantidade]
        return palavras_mais_frequentes

#Aplicando em Alice.txt:
caminho_arquivo = '/content/drive/MyDrive/Alice.txt'
processador = ProcessadorTexto(caminho_arquivo)

processador.abrir_arquivo()
total_palavras = processador.contar_palavras()
print(f'O número total de palavras no e-book é: {total_palavras}')

frequencia_palavras = processador.calcular_frequencia_palavras()
print('Frequência de cada palavra:')
for palavra, ocorrencias in frequencia_palavras.items():
    print(f'A palavra "{palavra}" tem frequência de {ocorrencias / total_palavras * 100:.2f}%')

quantidade_palavras_mais_frequentes = 20
palavras_mais_frequentes = processador.obter_palavras_mais_frequentes(quantidade_palavras_mais_frequentes)
print('As 20 palavras mais frequentes no e-book são:')
for palavra in palavras_mais_frequentes:
    frequencia = frequencia_palavras[palavra]
    print(f'Palavra: "{palavra}", Frequência: {frequencia / total_palavras * 100:.2f}%')
```
