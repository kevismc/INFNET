```python
# A Calabouço Discos é uma loja de vinis que resiste ao digital com albuns clássicos de lendas do rock. Os albuns são catalogados por: id, ano, artista, genero, titulo, preço. Os clientes são armazenados com: id, nome e idade. E as vendas da loja são armazenadas da seguinte maneira: id, id_cliente, ids_albuns (que é uma lista de IDs com os albuns que a pessoa comprou), total_da_venda.

# Utilizando dicionários, crie um código para registrar as vendas e ao final apresentar o fechamento do dia, apresentando as vendas uma a uma.
# Dicionário dos álbuns

catalogo_albuns = {
    1: {"ano": 1971, "artista": "Led Zeppelin", "genero": "Rock", "titulo": "Led Zeppelin IV", "preco": 50.0},
    2: {"ano": 1969, "artista": "The Beatles", "genero": "Pop", "titulo": "Abbey Road", "preco": 45.0},
    3: {"ano": 1973, "artista": "Pink Floyd", "genero": "Rock", "titulo": "The Dark Side of the Moon", "preco": 55.0},
    # Adicione mais álbuns aqui, se desejar
}

# Dicionário dos clientes
clientes = {
    1: {"nome": "João", "idade": 25},
    2: {"nome": "Maria", "idade": 50},
    3: {"nome": "Pedro", "idade": 60},
    # Adicione mais clientes aqui, se desejar
}

# Lista das vendas
vendas = []
# Dicionários para rastrear as contagens de gêneros em cada faixa etária
generos_18_25 = {"Rock": 0, "Pop": 0, "Outro": 0}
generos_25_35 = {"Rock": 0, "Pop": 0, "Outro": 0}
generos_35_55 = {"Rock": 0, "Pop": 0, "Outro": 0}
generos_acima_55 = {"Rock": 0, "Pop": 0, "Outro": 0}


# Função para registrar uma venda
def registrar_venda(id_cliente, ids_albuns):
    total_da_venda = 0
    for album_id in ids_albuns:
        album = catalogo_albuns.get(album_id)
        if album:
            total_da_venda += album["preco"]
        else:
            print(f"O álbum com ID {album_id} não foi encontrado.")
    
    cliente = clientes.get(id_cliente)
    if not cliente:
        print(f"O cliente com ID {id_cliente} não foi encontrado.")
        return
    
    venda = {"id_cliente": id_cliente, "ids_albuns": ids_albuns, "total_da_venda": total_da_venda}
    vendas.append(venda)
    
    
    generos_vendidos = set()
    for genero in generos_vendidos:
        idade_cliente = cliente["idade"]
        if 18 <= idade_cliente <= 25:
            generos_18_25[genero] += 1
        elif 25 <= idade_cliente <= 35:
            generos_25_35[genero] += 1
        elif 35 <= idade_cliente <= 55:
            generos_35_55[genero] += 1
        elif idade_cliente > 55:
            generos_acima_55[genero] += 1


# Solicitar entrada do usuário para adicionar uma venda
def adicionar_venda():
    print("===== Adicionar Venda =====")
    id_cliente = int(input("Digite o ID do cliente: "))
    ids_albuns = []
    while True:
        album_id = int(input("Digite o ID do álbum (ou 0 para encerrar): "))
        if album_id == 0:
            break
        ids_albuns.append(album_id)
    registrar_venda(id_cliente, ids_albuns)

# Registra as vendas
adicionar_venda()
adicionar_venda()

# Apresenta o fechamento do dia
print("Fechamento do dia:")
for venda in vendas:
    cliente = clientes.get(venda["id_cliente"])
    ids_albuns = venda["ids_albuns"]
    total_da_venda = venda["total_da_venda"]
    
    print(f"Cliente: {cliente['nome']}")
    print("Álbuns:")
    for album_id in ids_albuns:
        album = catalogo_albuns.get(album_id)
        if album:
            print(f"- {album['titulo']} (Artista: {album['artista']}, Preço: R${album['preco']})")
    print(f"Total da venda: R${total_da_venda}")
    print()

#Faturamento do dia:
num_vendas = len(vendas)
faturamento_total = sum(venda["total_da_venda"] for venda in vendas)


print(f"Total de vendas realizadas no dia: {num_vendas}")
print(f"Faturamento total da loja no dia: R${faturamento_total}")
#Exibe a maior compra e o cliente que mais gastou:
maior_compra = max(vendas, key=lambda venda: venda["total_da_venda"])
cliente_maior_gasto = clientes.get(maior_compra["id_cliente"])

print("Maior compra:")
print(f"Cliente: {cliente_maior_gasto['nome']}")
print(f"Valor da compra: R${maior_compra['total_da_venda']}")

#Exibir o album mais vendido:

contagem_albums = {}

for venda in vendas:
    ids_albuns = venda["ids_albuns"]
    for album_id in ids_albuns:
        if album_id in contagem_albums:
            contagem_albums[album_id] += 1
        else:
            contagem_albums[album_id] = 1
album_mais_vendido_id = max(contagem_albums, key=contagem_albums.get)
album_mais_vendido = catalogo_albuns.get(album_mais_vendido_id)

print("Álbum mais vendido:")
print(f"Título: {album_mais_vendido['titulo']}")
print(f"Artista: {album_mais_vendido['artista']}")
print(f"Total de vendas: {contagem_albums[album_mais_vendido_id]}")

#exibir por faixa etaria:
print("Gênero mais vendido por faixa etária:")
print("18 a 25 anos:", max(generos_18_25, key=generos_18_25.get))
print("25 a 35 anos:", max(generos_25_35, key=generos_25_35.get))
print("35 a 55 anos:", max(generos_35_55, key=generos_35_55.get))
print("Acima de 55 anos:", max(generos_acima_55, key=generos_acima_55.get))



```
