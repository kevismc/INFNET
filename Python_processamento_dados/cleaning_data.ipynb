### Tratamento da coluna role

```python
def tratamento_coluna (data:pd.DataFrame, serie:str, get:str) :
    
    series = list(data[serie])
    new_list = list()
    for elemento in series :
        new_list.append(eval(elemento).get(get))
    data[serie] = new_list

    return data
```

```python
data = tratamento_coluna(data=data, serie="role", get="S")
```

```python
data = tratamento_coluna(data=data, serie="fit", get="N")
```

```python
data = tratamento_coluna(data=data, serie="compressed", get="BOOL")
```

```python
data = tratamento_coluna(data=data, serie="over_time", get="BOOL")
```

```python
series = data["natural_percentage"].apply( lambda x: eval(x))
```

```python
dic = {
    "F" : [],
    "L" : [],
    "O" : [],
    "W" : []
}
for line in list(series) : 
    list_ = line["S"]
    dict_ = eval(list_)[0]

    for key, values in dict_.items() :
        dic[key].append(values)
    
for key, value in dic.items() :
    data[f"natural_{key.lower()}"] = value
```
